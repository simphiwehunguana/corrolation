# CATAGORICAL COLUNMS

```python
df1= pd.get_dummies(df,columns=["Education Status","Country"],dtype=int,drop_first=True)
```

```python
#I imporeted labelEncoder from sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder= LabelEncoder()
```

```python
#i created i new variable that i will be using onwards
#i did labelEncoding to the gender feature colunm
df1['Gender']=label_encoder.fit_transform(df1['Gender'])
```

# CORROLATION

```python
#I'm Analyzing correlation in table form so that it wiil easy to compare and identify the relationships without the need to examine individual scatter plots or visualizations
df1.corr() 
```

```python
# i Visualized the heatmap correlation of the variables between each other
# and i have observed that the relationships between each other ore not strongly corrolated.
plt.figure(figsize=(50,30))
correlation_matrix = df1.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='BuGn')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#I will calculate the Pearson correlation coefficient between the target variable "Wear Specs" and the other variables in the dataset.
wear_specs_correlations = df1.corr()['Wear Specs'].sort_values(ascending=False)
print("Correlation Coefficients with 'Wear Specs':")
print(wear_specs_correlations)
```

# I   created other advanced visualizations to identify trends and patterns

```python
#I created scatter plots to visualize the relationships between "Wear Specs" and the variables with high correlation coefficients.

plt.figure(figsize=(4, 3))
plt.scatter(df[ 'Time spent watching videos/TV'], df['Wear Specs'])
plt.xlabel( 'Time spent watching videos/TV')
plt.ylabel('Wear Specs')
plt.title('Time spent watching videos/TV vs Wear Specs')
plt.show()

plt.figure(figsize=(3, 3))
plt.scatter(df1['Age'], df['Wear Specs'])
plt.xlabel('Age')
plt.ylabel('Wear Specs')
plt.title( 'Age vs Wear Specs')
plt.show()

plt.figure(figsize=(4, 3))
plt.scatter(df1['Drinks alcohol'], df['Wear Specs'])
plt.xlabel('Drinks alcohol')
plt.ylabel('Wear Specs')
plt.title(' Drinks alcohol  vs Wear Specs')
plt.show()

plt.figure(figsize=(4, 3))
plt.scatter(df1['Whether parents have specs'], df['Wear Specs'])
plt.xlabel('Whether parents have specs')
plt.ylabel('Wear Specs')
plt.title('Whether parents have specs vs Wear Specs')
plt.show()
```

```python
#I created bar plots to visualize the distribution of the target variable across different categories of variables.
plt.figure(figsize=(4, 3))
df1.groupby('Gender')['Wear Specs'].mean().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Wear Specs')
plt.title('Bar Plot: Gender vs Wear Specs')
plt.show()

plt.figure(figsize=(4, 3))
df1.groupby( 'Maritial Status (0 - Single, 1 - Married, 2 - Divorced)')['Wear Specs'].mean().plot(kind='bar')
plt.xlabel( 'Maritial Status (0 - Single, 1 - Married, 2 - Divorced)')
plt.ylabel('Wear Specs')
plt.title(' Marital Status vs Wear Specs')
plt.show()

plt.figure(figsize=(4, 3))
df1.groupby('Education Status_Graduate')['Wear Specs'].mean().plot(kind='bar')
plt.xlabel('Education Status')
plt.ylabel('Wear Specs')
plt.title('Education Status_Graduate vs Wear Specs')
plt.show()
```

```python
# I created a histogram to visualize the distribution of the "IQ" variable.
plt.figure(figsize=(5, 4))
df1['IQ'].hist(bins=20)
plt.xlabel('IQ')
plt.ylabel('Count')
plt.title('Histogram: IQ')
plt.show()
```
